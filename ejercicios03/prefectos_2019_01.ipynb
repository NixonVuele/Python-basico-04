```python
# generar el proceso para prefectos 
# importacion de la libreria pandas
import pandas as pd
# lectura del archivo.sav con read_spss perteneciente a pandas (leemos el archivo.sav y lo guarmdamos en un dataframe )
datos_prefecto = pd.read_spss('data/resultadosdignidadprefecto2019.sav')
# presentacion del dataframe
datos_prefecto
```

```python
# lectura de del archivo candidatos2019.sav/ es necesaria esta lectura, debido a que contienen los nombres de los prefectos
datos_candidatos = pd.read_spss('data/candidatos2019.sav')
#  se muestra una cabezera del dataframe datos_candidatos
datos_candidatos.head()
```

```python
# aplicamos un merge, con el objetivo de hacer un solo dataframe que contenda el codigo,
# nombre del candidatos,op codigo y toda las informacion de los prefectos
dataframePrefectos = pd.merge(datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']], datos_prefecto, on="CANDIDATO_CODIGO")
# se muestra la cabezera del dataframePrefectos
dataframePrefectos.head()
```

```python
# el dataframe lo convertimos en csv con el .to_csv y ingresamos la ruta y el nombre del archivo .csv
dataframePrefectos.to_csv("data/datos_prefectos_2019_modificado.csv",index=False)
```

```python
# leemos el csv con read_csv he ingresamos la ruta donde se encuentra el archivo .csv y lo guardamos en el dataframe
lectura_prefectos = pd.read_csv("data/datos_prefectos_2019_modificado.csv")
# se muestra la cabezera del dataframe
lectura_prefectos.head()
```

```python
# permitir buscar por provincias:
# En la variable opciones se guarda una lista de tuplas, las que contiene los nombres de la provincia obtenidos del 
# dataframe lectura_prefectosy el .unique() permite que el nombre de la provincia no se repita dos veces el otra tupla
opciones = [(i,i) for i in lectura_prefectos.PROVINCIA_NOMBRE.unique()]
# opciones.sort permite ordenar la lista en orden alfabetico
opciones.sort()
# presenta opciones
print(opciones)
```

```python
# se importa la libreria iwydgets : La librería ipywidgets nos permite transformar un notebook de Jupyter 
# en un explorador de datos interactivo.
import ipywidgets as widgets

provincias = widgets.Dropdown(
    # options=[('AZUAY', 'AZUAY'), ('BOLÍVAR', 'BOLÍVAR'), ('CARCHI', 'CARCHI'), ('CAÑAR', 'CAÑAR'), ('CHIMBORAZO', 'CHIMBORAZO'), ('COTOPAXI', 'COTOPAXI'), ('EL ORO', 'EL ORO'), ('ESMERALDAS', 'ESMERALDAS'), ('GUAYAS', 'GUAYAS'), ('IMBABURA', 'IMBABURA'), ('LOJA', 'LOJA'), ('LOS RÍOS', 'LOS RÍOS'), ('MANABÍ', 'MANABÍ'), ('MORONA SANTIAGO', 'MORONA SANTIAGO'), ('NAPO', 'NAPO'), ('ORELLANA', 'ORELLANA'), ('PASTAZA', 'PASTAZA'), ('PICHINCHA', 'PICHINCHA'), ('SANTA ELENA', 'SANTA ELENA'), ('STO DGO TSÁCHILAS', 'STO DGO TSÁCHILAS'), ('SUCUMBÍOS', 'SUCUMBÍOS'), ('TUNGURAHUA', 'TUNGURAHUA'), ('ZAMORA CHINCHIPE', 'ZAMORA CHINCHIPE')]
    options = opciones,
    value= "AZUAY",
    description='Provincia:',
)

# se presentan los widgets
provincias
```

```python
# en datosPrefectos se guarda todas las filas que cumplan la condicion de que el nombre de la provincia sea igual 
# a la provincia seleccionada con el widget
datos_x_prefectos = lectura_prefectos[lectura_prefectos['PROVINCIA_NOMBRE']== provincias.value]

# a los datos_x_prefectos se los agrupa por el nombre de los candidatos; luego hace una suma de todos los votos dirigigos a los
# candidatos prefectos y los ordena al numero de votos de mayor a menor  y para finalizar lo presenta en un diagrama de barras

datos_x_prefectos.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="bar")
```
